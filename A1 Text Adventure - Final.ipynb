```python
"""
The following text adventure game was created based on the Harry Potter and the
Sorcerer's Stone movie. The game consists of getting through three stages to help
Harry reach the room where he will encounter Professor Quirrell, which 
is possessed by Lord Voldemort, and recover the Sorcerer Stone which can make a person immortal. 
In the game you can lose in the second and third stage
and it does require some previous knowledge of the movie to be able to win.

    First stage: Passing through the Devil's Snare
    Second stage: Getting the flying key to open the door
    Third stage: Playing human chess to pass to the next room
    
Bugs and/or Errors 

"""

import random

#defining the start of the game
def start():
    
    #printing the welcome of the game
    print("Welcome to the Harry Potter and the Sorcerer's Stone adventure!")
    print("Input your username to start playing")
    
    #prompting for username
    user_name = input(prompt = " ")
    
    #applying the username
    print(f"Welcome {user_name.capitalize()}, let's start! Your house is:")
    
    def house(hphouse):
        if hphouse == 1: 
            return('Gryffindor')
        elif hphouse == 2:
            return('Hufflepuff')
        elif hphouse == 3:
            return('Ravenclaw')
        elif hphouse == 4:
            return('Slytherin')
    print(house(random.randint(1,4)))
    
    #printing the introduction to the game
    print("""
    You're searching for the Sorcerer's Stone with Harry Potter, Ron and Hermione, you've just 
    encountered the three headed dog and played some music to make them fall asleep. 
    You all suddenly notice they are waking up, so you immediately jump inside
    the trapdoor and fall in something hard. But wait, it's getting all over your body...
    Hermione: 'It's the Devil's snare! We need to set us free or it will kill us!'
    """)

    input(prompt = "Press enter to continue")
    #the user goes into the first stage 
    stage_1()


# defining stage_1

def stage_1(): 
# printing the next questions so the user can choose one and try again if it's not correct
    print("""
    What would you do to escape the devil's snare?
        1) Try to move to liberate yourself
        2) Bite the snare 
        3) Stay still 
        """)

    choice_1 = input(prompt = "> ")
# converting charachters and letters to lower case
    choice_1 = choice_1.casefold()
    # creating a conditional statement
    #the third choice is the correct one so it will move the user to the next stage
    if choice_1 == "3" or "st" in choice_1:
        print("Great! Hermione, Harry and you just got free, you just need to help Ron get away too!")
        count = 3
        while count > 0:
            print(count)
            count -= 1
        input(prompt = "Press enter to continue\n")
        # to do the bold letter I searched it online
        print("\033[1m" + "LUMUS SOLEM!" + "\033[0m")
        stage_2()
    # conditional statement to try again    
    elif "1" in choice_1 or "m" in choice_1 or "liberate" in choice_1 or "try" in choice_1:
        print("No! Every time you move it gets more tangled around you! Think of something else")
        input(prompt = "Press any key to try again\n")
        stage_1()
    # conditional statement to try again 
    elif "2" in choice_1 or "bite" in choice_1 or "snare" in choice_1:
        print("Sorry, you are not able to pass through, try again\n")
        stage_1()
    # returning to the first stage     
    else:
        print("Invalid answer, please choose 1, 2 or 3")
        stage_1()

# the user goes into the second stage

# defining stage 2 
def stage_2():
    print("""
    That was a great spell to set him free, the Devil's Snare hates light! Now that you've  
    completed this obstacle, you can hear a noise in the room next to you, you open the door 
    and see thousands of fast flying keys. There's a door across the room, and next to you, a broom...
    
    You must answer the following questions to continue.""")
    
    question_list = ['Should I ride the broom and make an "Alohomora" spell?',
                     'Should I run away and find another way out?',]
    questions = 1
    
    # two possible random questions when playing and after that, one that must be answered 
    # creating a while loop
    while questions > 0:
        question = random.choice(question_list)
        input(prompt = "<Answer the following questions with 1 or 2, press enter to continue>\n")    
        print(question) 
       
    # creating a conditional statement
        if question == question_list[0]:
            print("1) I'll be able to open the door with Alohomora") 
            print("2) I need to find another way")
            questions -= 1
            #prompting for the answer
            answer = input(prompt = "> ")
            answer = answer.casefold()
            # creating a conditional statement 
            if answer == '1' or 'Al' in answer or 'do' in answer or 'op' in answer:
                print("Hermione: No Harry! The door is enchanted and we can't use a spell, let's keep trying\n")
                break
            elif answer == '2' or 'f' in answer or 'an' in answer or 'ne' in answer or 'w' in answer:
                print("Yes, I think we should\n")
                break
            else:
                print("Invalid answer, please choose 1 or 2")
                continue 
                           
        elif question == question_list[1]:
            print("1) There's no way out, we need to figure out a way to cross")
            print("2) I'll run until I found someone")
            questions -= 1
            #prompting for the answer
            answer = input(prompt = "> ")
            # converting charachters and letters to lower case
            answer = answer.casefold()
            
            if answer == '1' or 'w' in answer or 'th' in answer or 'cr' in answer:
                print("Ron: 'We are going to die in this place! What should we do?'\n")
                break               
            elif answer == '2' or 'run' in answer or 'fo' in answer or 'so' in answer:
                print("Hermione:'There is no one near, we need to find another way'\n") 
                break
            else: 
                print("Invalid answer, please try again") 
                continue
                
        # defining else clause so the loop can break           
        else:
            print("Invalid answer") 
            stage_2()
            break
    
    # end of while loop    
            
    print("Should I ride the broom and find the key that opens the door?") 
    answer = input(prompt = "Answer 'yes' or 'no'\n")
# converting charachters and letters to lower case
    answer = answer.casefold()
    # creating a conditional statement        
    if "y" in answer:
        print("We've got the key! Now we can open the door, let's continue our quest\n")
        stage_3()
            
    elif "n" in answer:
        fail()
            
    else:
        print("Please answer 'yes' or 'no'")
    

#defining stage 3
def stage_3():
    
    print("""
    You've crossed the door and found a huge room with a black and white floor.
    
        Ron: 'It's a chessboard! We've got to play our way across the room'
        
    Ron is an expert in wizard's chess so he tells you that he can be in charge 
    of the game.
    
    What should you do?
     1) Charm the chess pieces
     2) Play to checkmate
     3) Return for help
     """)
    choice_3 = input(prompt = "> ")
    # converting charachters and letters to lower case
    choice_3 = choice_3.casefold()
    # creating a conditional statement
    if choice_3 == "2" or "pl" in choice_3 or "chec" in choice_3:
        print("You just won the chess game! You can pass through")
        input(prompt = "Press any key to continue\n")
        winner()
        
    elif choice_3 == "1" or "chess" in choice_3 or "charm" in choice_3 or "pi" in choice_3:
        print("There's no way you can make it pass the enchanted chessboard, it's full of magic")
        fail()

    elif choice_3 == "3" or "re" in choice_3 or "help" in choice_3:
        print("I wish it was that easy")
        fail()
    # returning to stage 3     
    else:
        print("Invalid answer, please choose 1, 2 or 3")
        stage_3()
    

# defining the end of the game, using Awesome from Script 6
def winner():
    print("""You passed all the chambers and now arrived to fight Voldemort!
                                                _ 
     /\                                        | |
    /  \__      _____  ___  ___  _ __ ___   ___| |
   / /\ \ \ /\ / / _ \/ __|/ _ \| '_ ` _ \ / _ \ |
  / ____ \ V  V /  __/\__ \ (_) | | | | | |  __/_|
 /_/    \_\_/\_/ \___||___/\___/|_| |_| |_|\___(_)
 
    See you on the next game""")
          
#defining the failure of the game

def fail():
    # printing game over from ASCII
    print("""You failed at recovering the Sorcerer's Stone! 
   _____                                            
  / ____|                                           
 | |  __  __ _ _ __ ___   ___    _____   _____ _ __ 
 | | |_ |/ _` | '_ ` _ \ / _ \  / _ \ \ / / _ \ '__|
 | |__| | (_| | | | | | |  __/ | (_) \ V /  __/ |   
  \_____|\__,_|_| |_| |_|\___|  \___/ \_/ \___|_|   
                                                    
    See you next time!""")
    
    
start()
```
